```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl 
```

```python
%matplotlib inline
```

```python
nameList = ["A","B","C","D","E","F"]
ageList = [10,20,30,40,50,60]
weightList = [70,75,80,90,100,44]
```

```python
ageListNp = np.array(ageList)
weightListnp = np.array(weightList)
nameListNp = np.array(nameList)
```

```python
plt.plot(ageListNp,weightListnp,"y")
plt.xlabel("age")
plt.ylabel("weight")
plt.title("age-weight graph")
plt.show()
```

```python
my_npList = np.linspace(1,20,30)
my_npList2 = my_npList**10
```

```python
plt.plot(my_npList2,my_npList,"r.")
```

```python
plt.subplot(1,2,1) # 1 row , 2 column , 1. graph
plt.plot(my_npList2,my_npList,"y.")
plt.subplot(1,2,2)
plt.plot(my_npList,my_npList2,"y+")
```

```python
fig = plt.figure()
```

```python
fig.add_axes([1,1,0.5,0.5])
```
